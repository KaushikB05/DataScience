```python
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
digits = load_digits().data
Y = load_digits().target
```

```python
#row1 = digits[0,:]
row1 = digits[100,:]
```

```python
row1.shape
```

```python
img = row1.reshape(8,8)
```

```python
plt.imshow(img)
```

```python
# We cannot Vizualise in 64 dimensions so using PCA

pca = PCA(n_components=2)
tsne = TSNE()
```

```python
df_pca = pca.fit_transform(digits)
df_tsne = tsne.fit_transform(digits)
```

```python
plt.scatter(df_pca[:,0], df_pca[:,1],c=Y)
plt.show()
```

```python
plt.scatter(df_tsne[:,0], df_tsne[:,1],c=Y)
plt.show()
```
